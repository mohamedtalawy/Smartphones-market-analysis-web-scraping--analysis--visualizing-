```python
#1st step importing required Libraries
import requests
from bs4 import BeautifulSoup
import csv
import string
from itertools import zip_longest


from selenium import webdriver
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
import time
```

```python
Model = []
Price = []
Description = []
Link = []
brand = []
Color = []
Brand = []
spec = []
Storage = []
Rating = []
page_num = 0 
while page_num <14:
    #2nd step to fetch the url
    result = requests.get(f"https://www.jumia.com.eg/mobile-phones/apple--huawei--oppo--samsung--xiaomi/?sort=highest-price&viewType=grid&page={page_num}#catalog-listing")
    #4th step create soup object to parse content
    src = result.content
    
    
    #4th step create soup object to parse content
    soup = BeautifulSoup(src, "lxml")
    
    
    #Extracting Price Of each smartphone
    prices= soup.find_all("div",{"class":"prc"})
    for i in range(len(prices)):
        Price.append(prices[i].text.replace('EGP ', ''))

    rate= soup.find_all("div" , {"class":"stars _s"})
    for i in range(len(rate)):
        Rating.append(rate[i].text.strip())
        
        
        
    repElemList = soup.find_all('a',{"class":"core"})
    #extracting important data from <a> tag of each phone
    
    for repElem in repElemList:
        #Extracting Model Name 
        names = repElem.get('data-name')
        Model.append(names)

        #Extracting Brand Name 
        brands = repElem.get('data-brand')
        Brand.append(brands)

        #Extracting smartphone link 
        Links = repElem.get('href')
        Link.append(Links)    

        
    #eliminating none & null values in each list 
    Model = [i for i in Model if i]
    Brand = [i for i in Brand if i]
    Price = [i for i in Price if i]
    Link = [i for i in Link if i] 

    
#     #Extracting Phone Color
#     for i in range(len(Model)):
#         Color.append(Model[i].rsplit("-",3)[-1].strip())

#     #Extracting Phone specs
#     for i in range(len(Model)):
#         spec.append(Model[i])
#         #.rsplit("-",3)[-2].strip()[5:]
        
        
    page_num+=1
    print(f"page {page_num} scraped")
```

```python
Storage=[]
Ram = []
chr = '/'
for i in range(len(Model)):
    if chr in Model[i]:
        Storage.append(Model[i].split('/')[0][-5:].strip().replace('GB', '').replace('gb', ''))
        Ram.append(Model[i].split('/')[1][:4].replace('GB', '').replace('gb', ''))
        
    else:
        try:
            Storage.append(Model[i].split('-')[1].strip().replace('GB', '').replace('gb', ''))
            Ram.append('none')
        except IndexError:
            Storage.append('None')
```

```python
for i in range(len(Link)):
    Link[i] = 'https://www.jumia.com.eg' + Link[i]
```

```python
RatingValue=[]
RatinCount=[]
Shipping=[]
driver = webdriver.Chrome('chromedriver.exe')
for k in Link:
    
    driver.get(k)
    driver.maximize_window()

    time.sleep(2)
    try:
        Rating=driver.find_element_by_xpath('//div[@class="stars _s _al"]')
        RatingValue.append(Rating.text.split(' ')[0])
    except:
        RatingValue.append("No Rating")
            
    try:   
        rcount=driver.find_element_by_xpath('//a[@class="-plxs _more"]')
        RatinCount.append(rcount.text)
    except:
        RatinCount.append("0")
    try:
        shiping=driver.find_element_by_xpath('//div[@class="-df -d-co -c-bet"]/h3[@class="-fw"]')
        Shipping.append("FREE")
    except:
        Shipping.append("Paid")
        
```

```python
#From Lists To Columns and converting it into CSV file.

file_list = [Brand,Model,Price,Color,Storage,Ram,Link,RatingValue,RatinCount,Shipping]
exported = zip_longest(*file_list)

with open("./Jumia.csv","w",newline='') as myfile:
    wr = csv.writer(myfile)
    #naming attributes
    wr.writerow(["Brand","Model","Price","Color","Storage","Ram","Link","Rating","Rating Count","Shipping"])
    wr.writerows(exported)
```
